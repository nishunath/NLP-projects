````python
from collections import Counter
````

````python
def extract_keywords(text,num=60):
  tokens = [tok for tok in text.split()]
  most_common_tokens = Counter(tokens).most_common(num)
  return dict(most_common_tokens)
````

````python
keyword_positive=extract_keywords(positive_docx)
keyword_positive
````

````python
def most_common_review_words(mydict):
  df01=pd.DataFrame(mydict.items(), columns=['token','count'])
  plt.figure(figsize=(20,10))
  sns.barplot(x='token',y='count', data= df01)
  plt.xticks(rotation=45)
  plt.show()
````

````python
most_common_review_words(keyword_positive)
````

````python
from wordcloud import WordCloud
````

````python
def plot_wordcloud(docx):
  myworldcloud = WordCloud().generate(docx)
  plt.figure(figsize=(20,10))
  plt.imshow(myworldcloud,interpolation='bilinear')
  plt.axis('off')
  plt.show()
````

````python
plot_wordcloud(positive_docx)
````

````python
###Machine learning
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB

##Vectorizer
from sklearn.feature_extraction.text import  CountVectorizer,TfidfVectorizer

#Metrics
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,plot_confusion_matrix
````

````python
##Building features
Xfeatures=df['clean_review']
ylabels=df['sentiment']
Xfeatures
````

````python
#vectorizer apply
cv=CountVectorizer()
X=cv.fit_transform(Xfeatures)
````

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,ylabels,test_size=0.3)
````

````python
nb_model=MultinomialNB()
nb_model.fit(X_train,y_train)
````

````python
nb_model.score(X_test,y_test)
````

````python
nb_model_y_pred=nb_model.predict(X_test)
nb_model_y_pred
````

````python
#SAMPLE TEST
sample_text=['I watch the best film']
vect=cv.transform(sample_text).toarray()
nb_model.predict(vect)
````

````python
nb_model.classes_
````

````python
nb_model.predict_proba(vect)
````

````python
confusion_matrix(y_test,nb_model_y_pred)
````

````python
plot_confusion_matrix(nb_model,X_test,y_test)
````

````python
lr_model=LogisticRegression()
lr_model.fit(X_train,y_train)

lr_model.score(X_test,y_test)
````
